```python
import pandas as pd
import numpy as np
```

## reading words to be changed 

```python
f1=open("find_words.txt")
words_to_replace=f1.read().split()
f1.close()
```

```python
#words_to_replace
```

```python
dict_words = pd.read_csv('french_dictionary.csv', header=None, index_col=0, squeeze=True).to_dict()
```

## reading t8.shakespeare.txt

```python
f2=open("t8.shakespeare.txt")
book=f2.read()
f2.close()
```

```python
#book
```

## here we replaced and calculated time and memory

```python
import time
import tracemalloc

tracemalloc.start()
start = time.time()


d=dict()
for x in words_to_replace:
    d[x]=book.count(x)
    
    # here words are replaced with handling casing
    
    book=book.replace(x,dict_words[x])
    
    book=book.replace(x.upper(),dict_words[x])
    
    book=book.replace(x.title(),dict_words[x])
    
end = time.time()  

print("The output is memory in form of (current,peak)",tracemalloc.get_traced_memory())

tracemalloc.stop()
```

```python
f=open("performance.txt","w")
s="The output is memory in form of (current,peak)"+str(tracemalloc.get_traced_memory())
f.write(s)
s= "\nThe time of execution of above program is :"+str( end-start)                                                     
f.write(s)
f.close()
        
print("The time of execution of above program is :", end-start)
```
